# Typing Speed Test

A simple application to measure and improve your typing speed.

## Overview

This application allows users to test their typing speed by providing a standard text to type and measuring various metrics including words per second (WPS) and characters per minute (CPM). The application features a one-minute countdown timer and real-time feedback on typing performance.

## Features

- **Real-time Speed Measurement**: Tracks typing speed in words per second and characters per minute
- **Timed Tests**: Standard one-minute typing test
- **Interactive UI**: Visual feedback on performance with clear interface
- **Practice Text**: Standardized text for consistent practice

## How It Works

1. The application displays a sample text for the user to type
2. Timer starts automatically when the user begins typing
3. As the user types, the application calculates:
   - Words per second (WPS)
   - Characters per minute (CPM)
   - Elapsed time
   - Remaining time
4. The test ends after one minute, displaying final statistics

## Getting Started

### Prerequisites

- Python 3.6 or higher

### Installation

1. Clone this repository:
   ```
   git clone https://github.com/yourusername/typing-speed-test.git
   cd typing-speed-test
   ```

2. Run the application:
   ```
   python main.py
   ```

## Usage

1. Start the application
2. The sample text will be displayed on the screen
3. Begin typing the text in the input field
4. The timer will automatically start when you begin typing
5. Press space after each word
6. After one minute, your typing statistics will be displayed
7. Use the reset button to start a new test

## Code Structure

- `Header` class: Manages timing, statistics calculation, and UI components
- Sample text: Standard paragraph for typing practice
- Timer functionality: Tracks elapsed and remaining time
- Statistics calculation: Computes WPS and CPM based on user input
